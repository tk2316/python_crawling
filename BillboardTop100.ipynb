```python
import warnings
warnings.filterwarnings(action = 'ignore')
import requests
from bs4 import BeautifulSoup
```

```python
targetWebsite = "https://www.billboard.com/charts/hot-100"
header = {"User-agent": "Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko"}
request = requests.get(targetWebsite, headers = header)
#print(request)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
titles = soup.findAll('span', {'class':'chart-element__information__song text--truncate color--primary'})
artists = soup.findAll('span', {"class":"chart-element__information__artist text--truncate color--secondary"})
data = []
artist_data = []

for title in titles:
    data.append(title.text)
    
for artist in artists:
    artist_data.append(artist.text)
    
for i in range(0,len(data)):
    print('Rank: {0:3d}      Title:  {1}'.format(i+1, data[i]))
    print('               Artist: {}'.format(artist_data[i]))
    
```
