```python
import warnings
warnings.filterwarnings(action ='ignore')
import requests
from bs4 import BeautifulSoup          #bs4 페키지에 있는 BeautifulSoup 이라는 클라스
from datetime import datetime as dt   
```

```python
print(dt.now())
```

```python
targetWebsite = "https://music.bugs.co.kr/chart"
request = requests.get(targetWebsite)
html = request.text
soup = BeautifulSoup(html, 'html.parser')

titles = soup.findAll('p', {'class': 'title'})
for i in range(len(titles)):
    
    #two ways of doing the same crawling methods
    #1) print('{0:3d}위 {1}'.format(i+1, titles[i].text.strip()))      #strip() 공백을 제거해주는 함수다
    #2) print('{0:3d}위 {1}'.format(i+1, titles[i].text.split('\n')[1]))
    #print(titles[i].text.strip().split('\n')[0])
    print('{0:3d}위 {1}'.format(i+1, titles[i].text.strip().split('\n')[0]))
```

```python
#아티스트 검색
artists = soup.findAll('p', {'class': 'artist'})
for i in range(len(artists)):
    print('{0:3d}위 {1}'.format(i+1, artists[i].text.strip().split('\n')[0]))
    
```

```python
print('{} 현재'.format(dt.now()))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{0:3d}위 {1} - {2}'.format(i+1, artist, title))
```

```python
#크롤링한 결과를 텍스트 파일로 저장
#file = open('./output/bugsTop100.txt', 'w')
#위의 문장처럼 실행했는데 cp949 어쩌고 하면서 에러가 발생되면 open() 함수를 아래와 같이 수정한다
file = open('./output/bugsTop100.txt', 'w', -1, 'UTF-8')
file.write('{0} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{0:3d}위 {1} - {2}\n'.format(i+1, artist, title))
file.close()
print('bugTOP100.txt 파일로 쏘기 완료')
```

```python
#택스트 파일에 저장된 데이터 읽어서 화면에 출력하기
try:
    file = open('./output/bugsTop100.txt', 'r', -1, 'UTF-8')
    lines = file.readlines()
    for line in lines:
        print(line.strip())
    file.close()
    print('bugsTOP100.txt 파일에서 읽기 완료')
except FileNotFoundError:
    print('디스크에 bugsTOP100.txt 파일이 없습니다')
```
