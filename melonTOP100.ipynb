```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
request = requests.get(targetSite)
#406(허용되지 않음):요청한 페이지가 요청한 콘테츠 특성으로 응답할 수 어다.
print(request)     # <response [406]>
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
header = {"User-agent": "Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko"}
request = requests.get(targetSite,headers=header)
#print(request)     <Response [200]>
songsList = []
artistsList = []
html = request.text 
soup = BeautifulSoup(html, 'html.parser')

titles = soup.findAll('div',{'class':'ellipsis rank01'})
for i in range(len(titles)):
    title = ('{0:3d}위 {1}'.format(i+1,titles[i].text.strip()))
    songsList.append(title.strip())
```

```python
artists = soup.findAll('span',{'class':'checkEllipsis'})
for i in range(len(artists)):
    #print(artists[i].text.strip())
    artist = ('{0}'.format(artists[i].text.strip()))
    artistsList.append(artist)
```

```python
print('{0} 현재 melonTOP100'.format(dt.now()))
for i in range(len(artistsList)):
    print('{0} => {1}'.format(songsList[i], artistsList[i]))
```

```python
file = open('./output/melonTOP100.txt', 'w', -1, 'UTF-8')

targetSite = 'https://www.melon.com/chart/index.htm'
header = {"User-agent": "Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko"}
request = requests.get(targetSite,headers=header)
#print(request)     <Response [200]>
songsList = []
artistsList = []
html = request.text 
soup = BeautifulSoup(html, 'html.parser')

titles = soup.findAll('div',{'class':'ellipsis rank01'})
for i in range(len(titles)):
    title = ('{0:3d}위 {1}'.format(i+1,titles[i].text.strip()))
    songsList.append(title.strip())
    
artists = soup.findAll('span',{'class':'checkEllipsis'})
for i in range(len(artists)):
    #print(artists[i].text.strip())
    artist = ('{0}'.format(artists[i].text.strip()))
    artistsList.append(artist)

file.write('{0} 현재 melonTOP100\n'.format(dt.now()))
for i in range(len(artistsList)):
    file.write('{0} => {1} \n'.format(songsList[i], artistsList[i]))

file.close()
```
